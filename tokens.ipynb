```python
import tiktoken
encoding = tiktoken.encoding_for_model("gpt-5-nano")
token= encoding.encode("Hi i am LLM")
```

```python
token
```

```python
for token_id in token:
    token_text=encoding.decode([token_id])
    print(f"{token_text}= {token_id}")
```

```python
tokens=encoding.encode("LLM is the core part for AI, what's your opinion on this statement")
```
